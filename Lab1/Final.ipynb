```python
import numpy as np

# ==================================================
# 1. Gaussian probability density function
# ==================================================
def gaussian_pdf(x, mu, sigma):
    return (1 / np.sqrt(2 * np.pi * sigma**2)) * \
           np.exp(-0.5 * ((x - mu) / sigma)**2)

# ==================================================
# 2. Parameters from training (Sepal Length, x = 0)
# ==================================================
# Class w0: Setosa
mu0 = 5.01
sigma0 = 0.35

# Class w1: Versicolour
mu1 = 5.94
sigma1 = 0.52

# Prior probabilities
P0 = 0.5
P1 = 0.5

# ==================================================
# 3. Discriminant function g(x)
# ==================================================
def g(x):
    return gaussian_pdf(x, mu1, sigma1) * P1 - \
           gaussian_pdf(x, mu0, sigma0) * P0

# ==================================================
# 4. Posterior probabilities
# ==================================================
def posteriors(x):
    p0 = gaussian_pdf(x, mu0, sigma0) * P0
    p1 = gaussian_pdf(x, mu1, sigma1) * P1
    total = p0 + p1
    return p0 / total, p1 / total

# ==================================================
# 5. Classification with NO penalty (minimum error)
# ==================================================
def classify_no_penalty(x):
    return "w1" if g(x) > 0 else "w0"

# ==================================================
# 6. Risk-based classification (HIGH penalty)
# High penalty: misclassifying w0 as w1
# ==================================================
def classify_with_penalty(x, lambda_01=10):
    """
    lambda_01 = penalty for misclassifying Class 0 as Class 1
    """

    Pw0, Pw1 = posteriors(x)

    # Cost matrix:
    # Decide w0:
    #   correct w0 → cost 0
    #   w1 -> w0   → cost 1
    R0 = 0 * Pw0 + 1 * Pw1

    # Decide w1:
    #   w0 -> w1   → HIGH cost (lambda_01)
    #   correct w1 → cost 0
    R1 = lambda_01 * Pw0 + 0 * Pw1

    decision = "w0" if R0 < R1 else "w1"
    total_risk = min(R0, R1)

    return R0, R1, total_risk, decision

# ==================================================
# 7. COMPLETE PROGRAM (required by lab)
# ==================================================
def bayesian_classifier(x):
    Pw0, Pw1 = posteriors(x)
    gx = g(x)

    # No-penalty decision
    class_no_penalty = classify_no_penalty(x)

    # High-penalty decision + risks
    R0, R1, total_risk, class_penalty = classify_with_penalty(x)

    return {
        "x": x,
        "P(w0|x)": Pw0,
        "P(w1|x)": Pw1,
        "g(x)": gx,
        "Class (no penalty)": class_no_penalty,
        "Risk(decide w0)": R0,
        "Risk(decide w1)": R1,
        "Total Risk": total_risk,
        "Class (high penalty)": class_penalty
    }

# ==================================================
# 8. Demonstration (for report table)
# ==================================================
test_values = [3.3, 4.4, 1.2, 5.0, 5.7, 6.3, 1.5]

print("x\tP(w0|x)\tP(w1|x)\tg(x)\tNoPen\tR(w0)\tR(w1)\tTotRisk\tPenalty")
for x in test_values:
    r = bayesian_classifier(x)
    print(f"{x:.1f}\t"
          f"{r['P(w0|x)']:.3f}\t\t"
          f"{r['P(w1|x)']:.3f}\t\t"
          f"{r['g(x)']:.4f}\t"
          f"{r['Class (no penalty)']}\t"
          f"{r['Risk(decide w0)']:.3f}\t"
          f"{r['Risk(decide w1)']:.3f}\t"
          f"{r['Total Risk']:.3f}\t"
          f"{r['Class (high penalty)']}")
```
