```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.stats import norm


iris = load_iris()
X = iris.data          # features
y = iris.target        # labels

# Select Setosa (0) and Versicolour (1)
mask = y < 2
X = X[mask]
y = y[mask]

# Feature 0 = sepal length
x = X[:, 0]

x0 = x[y == 0]   # Class 0: Setosa
x1 = x[y == 1]   # Class 1: Versicolour

mu0, sigma0 = np.mean(x0), np.std(x0)
mu1, sigma1 = np.mean(x1), np.std(x1)

P0 = len(x0) / len(x)
P1 = len(x1) / len(x)

def g(x):
    return (np.log(norm.pdf(x, mu1, sigma1)) + np.log(P1)) - \
           (np.log(norm.pdf(x, mu0, sigma0)) + np.log(P0))
x_val = 4.5   # example value

if g(x_val) > 0:
    label = 1
else:
    label = 0



def classify(x):
    return 1 if g(x) > 0 else 0

def posteriors(x):
    p0 = norm.pdf(x, mu0, sigma0) * P0
    p1 = norm.pdf(x, mu1, sigma1) * P1
    total = p0 + p1
    return p0/total, p1/total

xs = np.linspace(min(x), max(x), 1000)
gx = g(xs)

idx = np.argmin(np.abs(gx))
mu1 = xs[idx]

plt.plot(xs, gx)
plt.axhline(0)
plt.axvline(mu1, color='r')
plt.title("Optimal Threshold μ₁")
plt.show()

lambda_10 = 5   # penalty for misclassifying w1 as w0

def g_penalty(x):
    return np.log(norm.pdf(x, mu1, sigma1)) + np.log(P1) - \
           (np.log(norm.pdf(x, mu0, sigma0)) + np.log(P0) + np.log(lambda_10))

test_x = [3.3, 4.4, 1.2, 5.0, 5.7, 6.3, 1.5]

for val in test_x:
    p0, p1 = posteriors(val)
    print(val, p0, p1, g(val), classify(val))

```
