```python
import numpy as np

# --------------------------------------------------
# Gaussian probability density function
# --------------------------------------------------
def gaussian_pdf(x, mu, sigma):
    return (1 / np.sqrt(2 * np.pi * sigma**2)) * \
           np.exp(-0.5 * ((x - mu) / sigma)**2)

# --------------------------------------------------
# Parameters obtained from training (sepal length)
# --------------------------------------------------
# Class w0: Setosa
mu0 = 5.01
sigma0 = 0.35

# Class w1: Versicolour
mu1 = 5.94
sigma1 = 0.52

# Prior probabilities
P0 = 0.5
P1 = 0.5

# --------------------------------------------------
# Discriminant function
# --------------------------------------------------
def g(x):
    return gaussian_pdf(x, mu1, sigma1) * P1 - \
           gaussian_pdf(x, mu0, sigma0) * P0

# --------------------------------------------------
# Posterior probabilities
# --------------------------------------------------
def posteriors(x):
    p0 = gaussian_pdf(x, mu0, sigma0) * P0
    p1 = gaussian_pdf(x, mu1, sigma1) * P1
    total = p0 + p1
    return p0 / total, p1 / total

# --------------------------------------------------
# Test feature values (given in lab manual)
# --------------------------------------------------
x_test = [3.3, 4.4, 1.2, 5.0, 5.7, 6.3, 1.5]

# --------------------------------------------------
# Build and display table
# --------------------------------------------------
print("x\tP(w0|x)\tP(w1|x)\tg(x)\tClass")
for x in x_test:
    Pw0, Pw1 = posteriors(x)
    gx = g(x)
    label = "w1" if gx > 0 else "w0"
    print(f"{x:.1f}\t{Pw0:.3f}\t\t{Pw1:.3f}\t\t{gx:.4f}\t{label}")
```
