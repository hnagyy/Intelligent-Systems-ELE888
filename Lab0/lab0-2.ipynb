```python
import numpy as np
import matplotlib.pyplot as plt

# Vectors and indexing
V = np.array([3, 6, 9, 1.25, -2])
v3 = V[2]
idx = np.where(V < 4)

# Matrices
M = np.array([[3, 4, 6],
              [2, 5, 7],
              [1, 2, 4]])

M_transpose = M.T          # M'
col = M[0:3, 1]
B = M[0:2, :]

K = np.array([[1, 2, 3],
              [4, 5, 6]])

L = np.array([[2, 3],
              [1, 2],
              [6, 5]])

P = K @ L                  # Matrix multiplication (K * L)
P2 = P @ P                 # P^2 (matrix power)

N = M[:, 0:2].T            # M(:,1:2)'
add_result = K + N         # K + N
sub_result = K - N         # K - N

Z = np.zeros((2, 4))
F = 4 * np.ones((1, 3))

diagM = np.diag(M)
I = np.eye(3, 3)

# ----- Sample feature space -----
A = np.array([0.1, 0.3, 0.25, 0.4, 0.15, 0.51, 0.62, 0.8, 0.58, 0.9])
B = np.array([0.01, 0.23, 0.5, 0.36, 0.27, 0.75, 0.68, 0.59, 0.65, 0.8])

plt.figure()
plt.plot(A[0:5], B[0:5], 'b+', label='Class 1')
plt.plot(A[5:10], B[5:10], 'ro', label='Class 2')

plt.axis([0, 1, 0, 1])
plt.xlabel('Sample Feature 1')
plt.ylabel('Sample Feature 2')
plt.title('Feature Space')
plt.legend()
plt.grid(True)

# ----- 3D surface plot -----
# meshgrid equivalent
X = np.arange(-3, 3.125, 0.125)
Y = np.arange(-3, 3.125, 0.125)
X, Y = np.meshgrid(X, Y)

# Peaks-like function (Matlab's peaks has no direct built-in in Python)
def peaks(x, y):
    return 3*(1-x)**2 * np.exp(-x**2 - (y+1)**2) \
           - 10*(x/5 - x**3 - y**5) * np.exp(-x**2 - y**2) \
           - 1/3*np.exp(-(x+1)**2 - y**2)

Z = peaks(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-10, 5])

ax.set_title('3D Surface Plot')
plt.show()


```
