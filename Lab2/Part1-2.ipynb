```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


def build_A_B_signed(seed=42):
    
    iris = load_iris()
    X = iris.data[:, [1, 2]]   # x1, x2
    y = iris.target            # class labels: 0,1,2

    A = X[y == 0]  # Setosa
    B = X[y == 1]  # Versicolour

    # Augment: y = [1, x1, x2]
    YA = np.hstack([np.ones((A.shape[0], 1)), A])   # (50,3)
    YB = np.hstack([np.ones((B.shape[0], 1)), B])   # (50,3)

    # Signed transform: class B samples are negated
    Y_all = np.vstack([YA, -YB])  # (100,3)

    # Shuffle
    rng = np.random.default_rng(seed)
    idx = rng.permutation(Y_all.shape[0])
    return Y_all[idx]
    
def build_B_C_signed(seed=42):

    iris = load_iris()
    X = iris.data[:, [1, 2]]   # x1, x2
    y = iris.target

    B = X[y == 1]  # Versicolour
    C = X[y == 2]  # Virginica

    YB = np.hstack([np.ones((B.shape[0], 1)), B])  # (50,3)
    YC = np.hstack([np.ones((C.shape[0], 1)), C])  # (50,3)

    
    Y_all= np.vstack([YB, -YC])  # (100,3)

    rng = np.random.default_rng(seed)
    idx = rng.permutation(Y_all.shape[0])
    return Y_all[idx]


def split_train_test(Y_all, train_frac=0.30):
    n = Y_all.shape[0]
    n_train = int(round(train_frac * n))
    Y_train = Y_all[:n_train]
    Y_test  = Y_all[n_train:]
    return Y_train, Y_test

def train_gd_perceptron(Y_train, eta=0.01, theta=0.0, a0=None, max_iter=300):
    if a0 is None:
        a = np.array([0.0, 0.0, 1.0], dtype=float)
    else:
        a = np.array(a0, dtype=float)

    history = []  # store (k, Jp, grad, a_copy, num_mis)

    for k in range(1, max_iter + 1):
        scores = Y_train @ a  #a^Tyi​
        mis = scores <= 0  # misclassified set Y(a)

        num_mis = int(np.sum(mis))
        if num_mis == 0:
            # exit if no misclassified
            break

        Ymis = Y_train[mis]

        grad = -np.sum(Ymis, axis=0)            # ∇Jp = Σ(-y)
        Jp = np.sum(-(Ymis @ a))                # Jp = Σ(-a^T y)

        history.append((k, float(Jp), grad.copy(), a.copy(), num_mis))

        # Update a value
        a = a - eta * grad

        # Optional stop rule from pseudo-code:
        if np.linalg.norm(eta * grad) < theta:
            break

    return a, history
    
def signed_to_xy_and_class(Y_signed):
    """
    Your signed samples look like:
      Class A:  [ 1,  x1,  x2]
      Class B:  [-1, -x1, -x2]
    So we can recover original (x1,x2) and class from the first entry.
    """
    y0 = Y_signed[:, 0]
    x1 = np.where(y0 > 0, Y_signed[:, 1], -Y_signed[:, 1])
    x2 = np.where(y0 > 0, Y_signed[:, 2], -Y_signed[:, 2])
    cls = np.where(y0 > 0, 0, 1)   # 0 = A, 1 = B
    return x1, x2, cls

def history_to_table(history):
    """
    history entries were stored as:
      (k, Jp, grad_vector, a_vector, num_mis)
    Task 5 wants: iteration, Jp(a), ∇Jp(a), a=[a0,a1,a2]
    :contentReference[oaicite:6]{index=6}
    """
    rows = []
    for (k, Jp, grad, a, num_mis) in history:
        rows.append({
            "k": k,
            "Jp(a)": Jp,
            "grad": grad,     # vector
            "a": a,           # vector
            "misclassified": num_mis
        })
    return pd.DataFrame(rows)

def plot_training_and_boundary(Y_train_signed, a, title=""):
    x1, x2, cls = signed_to_xy_and_class(Y_train_signed)

    # scatter training points
    plt.figure()
    plt.scatter(x1[cls == 0], x2[cls == 0], label="Class A")
    plt.scatter(x1[cls == 1], x2[cls == 1], label="Class B")

    # decision boundary: a0 + a1*x1 + a2*x2 = 0
    a0, a1, a2 = a
    x1_min, x1_max = x1.min(), x1.max()
    xs = np.linspace(x1_min, x1_max, 200)

    if abs(a2) < 1e-12:
        # vertical line if a2 ~ 0: a0 + a1*x1 = 0
        if abs(a1) > 1e-12:
            x_vert = -a0 / a1
            plt.axvline(x_vert, linestyle="--", label="Decision boundary")
    else:
        ys = -(a0 + a1 * xs) / a2
        plt.plot(xs, ys, linestyle="--", label="Decision boundary")

    plt.xlabel("x1 = sepal width")
    plt.ylabel("x2 = petal length")
    plt.title(title)
    plt.legend()
    plt.show()

def run_step5_AB():
 
    Y_all = build_A_B_signed(seed=7)
    Y_train, _ = split_train_test(Y_all, train_frac=0.70)  # choose 70% training for nicer plots

    etas = [0.01, 0.1]  # two different learning rates :contentReference[oaicite:7]{index=7}
    inits = [
        [0, 0, 1],
        [1, 1, 1],
        [-1, 0, 1],
    ]

    for eta in etas:
        for a0 in inits:
            a_final, hist = train_gd_perceptron(
                Y_train, eta=eta, theta=0.0, a0=a0, max_iter=300
            )

            # convergence iterations = last iteration in history, or 0 if already perfect
            iters = hist[-1][0] if len(hist) else 0

            print("\n==============================")
            print(f"eta = {eta}, a(0) = {a0}")
            print("final a =", a_final)
            print("iterations used =", iters)

            # 1) iteration table :contentReference[oaicite:8]{index=8}
            df = history_to_table(hist)
            print(df.head(10))  # show first 10 rows (you can print all if you want)

            # 2) training data + 3) decision boundary :contentReference[oaicite:9]{index=9}
            plot_training_and_boundary(
                Y_train, a_final,
                title=f"A vs B | eta={eta} | a0={a0} | iters={iters}"
            )

# Run Task 5 for A vs
# ==============================
# NEW: accuracy computation
# ==============================
def compute_accuracy(Y_test, a):
    """
    For signed samples:
      correctly classified if a^T y > 0
    """
    scores = Y_test @ a
    correct = scores > 0
    return np.mean(correct)


# ==============================
# NEW: run Tasks 1–4
# ==============================
def run_tasks_1_to_4():
    print("\n================ TASKS 1–4 =================")

    # ---- SETTINGS FROM QUESTION ----
    eta = 0.01
    theta = 0.0
    a_init = [0, 0, 1]
    max_iter = 300

    # helper to run one experiment
    def run_experiment(name, build_func, train_frac):
        print("\n--------------------------------")
        print(f"{name} | train={int(train_frac*100)}% "
              f"test={int((1-train_frac)*100)}%")

        # build dataset
        Y_all = build_func(seed=7)

        # split
        Y_train, Y_test = split_train_test(Y_all, train_frac=train_frac)

        # train
        a_final, hist = train_gd_perceptron(
            Y_train,
            eta=eta,
            theta=theta,
            a0=a_init,
            max_iter=max_iter
        )

        iters = hist[-1][0] if len(hist) else 0
        acc = compute_accuracy(Y_test, a_final)

        print("final a =", a_final)
        print("iterations =", iters)
        print(f"classification accuracy = {acc*100:.2f}%")

        # optional plot
        plot_training_and_boundary(
            Y_train,
            a_final,
            title=f"{name} | train={int(train_frac*100)}%"
        )

    # -------------------------------
    # TASK 1 + 2  (A vs B, 30/70)
    # -------------------------------
    run_experiment(
        "A vs B",
        build_A_B_signed,
        train_frac=0.30
    )

    # -------------------------------
    # TASK 3 (A vs B, 70/30)
    # -------------------------------
    run_experiment(
        "A vs B",
        build_A_B_signed,
        train_frac=0.70
    )

    # -------------------------------
    # TASK 4 (B vs C, both splits)
    # -------------------------------
    run_experiment(
        "B vs C",
        build_B_C_signed,
        train_frac=0.30
    )

    run_experiment(
        "B vs C",
        build_B_C_signed,
        train_frac=0.70
    )
# Run Tasks 1–4
run_tasks_1_to_4()

# Run Task 5
run_step5_AB()
```
