```python
import numpy as np
from sklearn.datasets import load_iris

# ----------------------------
# Part 1 (Task 1 + prints Task 2 accuracy)
# A vs B, 30% train / 70% test
# eta = 0.01, theta = 0, a(0) = [0,0,1], max_iter = 300
# ----------------------------

def build_A_B_signed(seed=42):
    
    iris = load_iris()
    X = iris.data[:, [1, 2]]   # x1, x2
    y = iris.target            # class labels: 0,1,2

    A = X[y == 0]  # Setosa
    B = X[y == 1]  # Versicolour

    # Augment: y = [1, x1, x2]
    YA = np.hstack([np.ones((A.shape[0], 1)), A])   # (50,3)
    YB = np.hstack([np.ones((B.shape[0], 1)), B])   # (50,3)

    # Signed transform: class B samples are negated
    Y_all = np.vstack([YA, -YB])  # (100,3)

    # Shuffle
    rng = np.random.default_rng(seed)
    idx = rng.permutation(Y_all.shape[0])
    return Y_all[idx]
    
def build_B_C_signed(seed=42):

    iris = load_iris()
    X = iris.data[:, [1, 2]]   # x1, x2
    y = iris.target

    B = X[y == 1]  # Versicolour
    C = X[y == 2]  # Virginica

    YB = np.hstack([np.ones((B.shape[0], 1)), B])  # (50,3)
    YC = np.hstack([np.ones((C.shape[0], 1)), C])  # (50,3)

    Y_all = np.vstack([YB, -YC])  # (100,3)

    rng = np.random.default_rng(seed)
    idx = rng.permutation(Y_all.shape[0])
    return Y_all[idx]


def split_train_test(Y_all, train_frac=0.30):
    n = Y_all.shape[0]
    n_train = int(round(train_frac * n))
    Y_train = Y_all[:n_train]
    Y_test  = Y_all[n_train:]
    return Y_train, Y_test

def train_gd_perceptron(Y_train, eta=0.01, theta=0.0, a0=None, max_iter=300):
    if a0 is None:
        a = np.array([0.0, 0.0, 1.0], dtype=float)
    else:
        a = np.array(a0, dtype=float)

    history = []  # store (k, Jp, grad, a_copy, num_mis)

    for k in range(1, max_iter + 1):
        scores = Y_train @ a  #a^Tyi​
        mis = scores <= 0  # misclassified set Y(a)

        num_mis = int(np.sum(mis))
        if num_mis == 0:
            # exit if no misclassified
            break

        Ymis = Y_train[mis]

        grad = -np.sum(Ymis, axis=0)            # ∇Jp = Σ(-y)
        Jp = np.sum(-(Ymis @ a))                # Jp = Σ(-a^T y)

        history.append((k, float(Jp), grad.copy(), a.copy(), num_mis))

        # Update a value
        a = a - eta * grad

        # Optional stop rule from pseudo-code:
        if np.linalg.norm(eta * grad) < theta:
            break

    return a, history

def test_accuracy_signed(Y_test, a):
    # test sample is correct if a^T y > 0.
    scores = Y_test @ a
    correct = scores > 0
    return 100.0 * np.mean(correct)

if __name__ == "__main__":

    # -----------------------------
    # A vs B (Setosa vs Versicolour)
    # -----------------------------
    print("========== A vs B ==========")
    Y_all_AB = build_A_B_signed(seed=7)

    # Part 1 + 2: 30% train / 70% test
    Y_train, Y_test = split_train_test(Y_all_AB, train_frac=0.30)
    a_AB_30, hist_AB_30 = train_gd_perceptron(
        Y_train, eta=0.01, theta=0.0, a0=[0, 0, 1], max_iter=300
    )
    acc_AB_70test = test_accuracy_signed(Y_test, a_AB_30)
    print("\nA vs B (30/70)")
    print("Final a =", a_AB_30)
    print("Iterations =", (hist_AB_30[-1][0] if len(hist_AB_30) else 0))
    print("Test accuracy (70% test) = {:.2f}%".format(acc_AB_70test))

    # Part 3: 70% train / 30% test
    Y_train, Y_test = split_train_test(Y_all_AB, train_frac=0.70)
    a_AB_70, hist_AB_70 = train_gd_perceptron(
        Y_train, eta=0.01, theta=0.0, a0=[0, 0, 1], max_iter=300
    )
    acc_AB_30test = test_accuracy_signed(Y_test, a_AB_70)
    print("\nA vs B (70/30)")
    print("Final a =", a_AB_70)
    print("Iterations =", (hist_AB_70[-1][0] if len(hist_AB_70) else 0))
    print("Test accuracy (30% test) = {:.2f}%".format(acc_AB_30test))

    # -----------------------------
    # B vs C (Versicolour vs Virginica)
    # -----------------------------
    print("\n========== B vs C ==========")
    Y_all_BC = build_B_C_signed(seed=7)

    # Part 4a: 30% train / 70% test
    Y_train, Y_test = split_train_test(Y_all_BC, train_frac=0.30)
    a_BC_30, hist_BC_30 = train_gd_perceptron(
        Y_train, eta=0.01, theta=0.0, a0=[0, 0, 1], max_iter=300
    )
    acc_BC_70test = test_accuracy_signed(Y_test, a_BC_30)
    print("\nB vs C (30/70)")
    print("Final a =", a_BC_30)
    print("Iterations =", (hist_BC_30[-1][0] if len(hist_BC_30) else 0))
    print("Test accuracy (70% test) = {:.2f}%".format(acc_BC_70test))

    # Part 4b: 70% train / 30% test
    Y_train, Y_test = split_train_test(Y_all_BC, train_frac=0.70)
    a_BC_70, hist_BC_70 = train_gd_perceptron(
        Y_train, eta=0.01, theta=0.0, a0=[0, 0, 1], max_iter=300
    )
    acc_BC_30test = test_accuracy_signed(Y_test, a_BC_70)
    print("\nB vs C (70/30)")
    print("Final a =", a_BC_70)
    print("Iterations =", (hist_BC_70[-1][0] if len(hist_BC_70) else 0))
    print("Test accuracy (30% test) = {:.2f}%".format(acc_BC_30test))
```
